```python
import numpy as np
import jumpdiff as jd
import matplotlib.pyplot as plt
```

```python
# integration time and time sampling
t_final = 10000
delta_t = 0.001

# A drift function
def a(x):
    return -0.5*x

# and a (constant) diffusion term
def b(x):
    return 0.2*abs(x)**2

# Now define a jump amplitude and rate
xi = 1
lamb = 1.75

# and simply call the integration function
X = jd.jd_process(t_final, delta_t, a=a, b=b, xi=xi, lamb=lamb, init = 0.1)
plt.plot(X)
```

```python
edges, moments = jd.moments(timeseries = X, bw = 0.4)
```

```python

drift = -0.5*edges
plt.plot(edges, moments[1]/delta_t, label = 'extracted moments')
plt.plot(edges,drift, label ='theoretical drift')
plt.xlim((-10,10))
plt.ylim((-10,10))
plt.legend()
```

```python
second_moment = 0.04*np.abs(edges)**4 + (xi)*lamb
plt.plot(edges, moments[2]/delta_t, label = 'extracted moments')
plt.plot(edges, second_moment, label ='theoretical drift')
plt.xlim((-20,20))
plt.ylim((0,100))
plt.legend()
```
