```python
from Models.product_log import AsymmetricDriftModel
from Models.piecewise import LinearPowerLaw
from Models.gamma import GammaDiffusion
import numpy as np


for site_id in results:
    try:
        edges = results[site_id]['KM_results']['edges']
        moments = results[site_id]['KM_results']['F1']
        if 'fit' not in results[site_id]:
            results[site_id]['fit'] = {}
        drift_model = AsymmetricDriftModel(lower_trim= 0.05, upper_trim=0.95, use_weights= False)
        drift_model.fit(edges, moments)
        y_fit = np.linspace(edges[0], edges[len(edges)-1])
        results[site_id]['fit']['F1'] = {
            'y_fit': y_fit,
            'model_fit': drift_model.evaluate(y_fit),
            'params': drift_model.params}

    except Exception as e:
        print(f"Skipping {site_id}: {e}")

for site_id in results:
    try:
        edges = results[site_id]['KM_results']['edges']
        moments = results[site_id]['KM_results']['F2_corrected']
        diffusion_model = GammaDiffusion()
        diffusion_model.fit(edges, moments)
        y_fit = np.linspace(edges[0], edges[len(edges)-1])
        results[site_id]['fit']['F2'] = {
            'y_fit': y_fit,
            'model_fit': diffusion_model.evaluate(y_fit),
            'params': diffusion_model.params}
    except Exception as e:
        print(f"Skipping {site_id}: {e}")
```

```python
def plot_parameter_histograms(results, bins=20, figsize_per_param=(4, 3), clip_percentile=95):

    def extract_params(results, moment, param_list):
        data = {p: [] for p in param_list}
        for s in results:
            fit = results[s].get('fit', {}).get(moment)
            if fit:
                for p in param_list:
                    if p in fit['params']:
                        data[p].append(fit['params'][p])
        return data

    def plot_params(data, param_list, title, color):
        n = len(param_list)
        fig, axes = plt.subplots(1, n, figsize=(n * figsize_per_param[0], figsize_per_param[1]))
        fig.suptitle(title, fontsize=12)
        for ax, p in zip(axes, param_list):
            vals = np.array(data[p])
            lo = np.percentile(vals, 100 - clip_percentile)
            hi = np.percentile(vals, clip_percentile)
            ax.hist(vals, bins=bins, color=color, edgecolor='white', range=(lo, hi))
            ax.set_title(p)
            ax.set_xlabel('Value')
            ax.set_ylabel('Count')
        fig.tight_layout()
        return fig

    f1_params = ['y_eq', 'm', 'a', 'alpha']
    f2_params = ['C', 'a', 'beta', 'gamma', 'y_peak']

    f1_data = extract_params(results, 'F1', f1_params)
    f2_data = extract_params(results, 'F2', f2_params)

    fig_f1 = plot_params(f1_data, f1_params, 'ProductModel (F1) Parameters', 'steelblue')
    fig_f2 = plot_params(f2_data, f2_params, 'GammaDiffusion (F2) Parameters', 'coral')

    return fig_f1, fig_f2


fig_f1, fig_f2 = plot_parameter_histograms(results)
```

```python
import jumpdiff as jd
def simulate_and_compare(results, site_id):
    
    f1_params = results[site_id]['fit']['F1']['params']
    f2_params = results[site_id]['fit']['F2']['params']
    timescale = results[site_id]['timescale']

    y_eq  = f1_params['y_eq']
    m     = f1_params['m']
    a     = f1_params['a']
    alpha = f1_params['alpha']
    
    C     = f2_params['C']
    a2    = f2_params['a']
    beta  = f2_params['beta']
    gamma = f2_params['gamma']
    
    def drift(x):
        diff = x - y_eq
        return m * diff * (1 + a * np.abs(diff)**(alpha - 1))
    
    def diffusion(x):
        x = np.maximum(x, 0)
        return np.sqrt(np.maximum(C + a2 * np.power(x, beta) * np.exp(-gamma * x), 0))
    
    delta_t = 15
    t_final = 300000
    print(f"Timescale: {timescale/60/24:.1f} days")
    print(f"Simulating: {t_final:.1f} timescale units ({int(t_final/delta_t)} steps)")
    
    X = jd.jd_process(
        t_final,
        float(delta_t),
        a=drift,
        b=diffusion,
        xi=0,
        lamb=0.0,
        init=float(1.1)
    )
    
    original = results[site_id]['df']['y_norm'].to_numpy()
    original = original[np.isfinite(original)]
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    fig.suptitle(f'Site {site_id}', fontsize=12)
    
    bins = np.linspace(0, 6, 50)
    axes[0].hist(original, bins=bins, density=True, alpha=0.6,
                 color='k', label='Observed')
    axes[0].hist(X[X > 0], bins=bins, density=True, alpha=0.6,
                 color='r', label='Simulated')
    axes[0].set_xlabel('y (normalized)')
    axes[0].set_ylabel('Density')
    axes[0].set_title('Marginal distribution')
    axes[0].legend()
    
    t = np.arange(len(X)) * delta_t
    axes[1].plot(t[:20000], X[:20000], 'r-', alpha=0.5, lw=0.5)
    axes[1].set_xlabel('t (timescale units)')
    axes[1].set_ylabel('y (normalized)')
    axes[1].set_title('Simulated trajectory')
    
    fig.tight_layout()
    return fig, X
```
